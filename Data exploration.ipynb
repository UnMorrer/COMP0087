```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel('C:/Users/tducr/Music/UCL/Term2/COMP0087/project/Data set/1/training_set_rel3.xls')
data=df.iloc[:,:3]
data['cnt']=data['essay'].str.split().str.len()
```

```python
viol=[]
for i in pd.unique(data['essay_set']):
    viol.append(data[data['essay_set']==i]['cnt'])
```

```python
plt.violinplot(viol,showmedians=True,quantiles=[[0.25,0.75],[0.25,0.75],[0.25,0.75],[0.25,0.75],[0.25,0.75],[0.25,0.75],[0.25,0.75],[0.25,0.75]])
plt.title('Number of words in the essays')
plt.xlabel('Questions')
plt.ylabel('Number of words')
plt.show()
```

```python
viol[0].count()
plt.bar(np.arange(1,len(viol)+1),[v.count() for v in viol])
plt.title('Number of essays for each questions')
plt.xlabel('Questions')
plt.ylabel('Number of Essays')
plt.show()
```
